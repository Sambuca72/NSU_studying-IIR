```python
# name_of_file = input("Введите имя файла с письмами: ")
import string
name_of_file = "email1.txt"
with open(name_of_file) as f:
    lines = f.readlines()
    emails = []

for s in lines:
    s = s.translate(str.maketrans("", "", string.punctuation)).lower()
    words = set(s.split())
    emails.append(words)
print(emails)
```
